```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
X=titanic.drop('Survived',axis=1)
y=titanic['Survived'].copy()
```

```python
X
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
reg=RandomForestClassifier()
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
txt=X_train.select_dtypes(exclude=[np.number]).columns
num=X_train.select_dtypes(include=[np.number]).columns
```

```python
txtpip=Pipeline([('impute',SimpleImputer(strategy='constant',fill_value='frequent')),
                 ('onehot',OneHotEncoder())])
numpip=Pipeline([('impute',SimpleImputer(strategy='median'))])
```

```python
transformer=ColumnTransformer([('onehot',txtpip,txt),
             ('numpip',numpip,num)],
             remainder='passthrough')
```

```python
titanic
```

```python
titanic.columns.tolist()
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
Xtt=transformer.fit_transform(X_train)
tc=transformer.get_feature_names_out()
Xttdf=pd.DataFrame(Xtt,columns=tc)
```

```python
Xtst=transformer.transform(X_test)
tc=transformer.get_feature_names_out()
Xtstdf=pd.DataFrame(Xtst,columns=tc)
```

```python
reg.fit(Xttdf,y_train)
```

```python
reg.score(Xtstdf,y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
full_pipeline1 = Pipeline([
('preprocessing', transformer),
('random_forest', model)])

full_pipeline1.fit(X_train, y_train)

preds1=full_pipeline1.predict(X_test)

preds1
```

```python
full_pipeline1.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()

param_grid={
    'n_estimators':[50,60],
    'max_depth':[10,15,30]

}

grid_search=GridSearchCV(estimator=model,param_grid=param_grid,cv=3)
grid_search.fit(Xttdf,y_train)
```

```python
grid_search.score(Xtstdf,y_test)
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()

param_grid ={
    'max_depth':[10,15,30],
    'max_features':['sqrt','log2'],
    'splitter':['best','random']
}

tree=DecisionTreeRegressor()
rig=GridSearchCV(tree,param_grid=param_grid)
rig.fit(Xttdf,y_train)
```

```python
rig.best_params_
```

```python
rig.score(Xtstdf,y_test)
```

```python
rig.best_estimator_.score(Xtstdf,y_test)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score

sgd_clf= SGDClassifier(random_state=42)
sgd_clf.fit(Xttdf,y_train)
```

```python
from sklearn.model_selection import cross_val_predict

y_train_pred=cross_val_predict(sgd_clf,Xttdf,y_train,cv=3)
y_train_pred
```

```python
from sklearn.metrics import precision_score, recall_score

precision_score(y_train,y_train_pred)
```

```python
recall_score(y_train,y_train_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_train,y_train_pred)
```
